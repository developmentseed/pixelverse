```python
SAMPLE_AOI_BBOX = (34.30401965358922, 0.483473285913126, 34.3178177763533, 0.497271408677202)
# Get items (lazily loaded)
items_by_month = get_s2_times_series(bbox=SAMPLE_AOI_BBOX, year=2023, cloudcover_max=60, months=[6])
```

```python
items_by_month 
```

```python
rgb = items_by_month[["red", "green", "blue"]].isel(time=0)
rgb_scaled = (255 * rgb / 3000).clip(0, 255).astype("uint8")
rgb_scaled.rio.to_raster('./s2_rgb_month_test.tif')
```

```python
# Display RGB composite scaled to 3000
import numpy as np
import matplotlib.pyplot as plt
rgb = items_by_month[["red", "green", "blue"]].isel(time=0)
rgb_array = np.stack([rgb["red"].values, rgb["green"].values, rgb["blue"].values], axis=-1)
rgb_scaled = np.clip(rgb_array / 3000, 0, 1)

plt.figure(figsize=(10, 10))
plt.imshow(rgb_scaled)
plt.axis('off')
plt.title('RGB Composite - Month 1')
plt.tight_layout()
plt.show()
```
